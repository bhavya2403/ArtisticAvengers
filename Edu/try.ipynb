```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./file.xlsx')

# print(df.head())

# create a new data frame with all the columns except 15 to 25

df1 = df.iloc[:, 0:5]

# add last 11 columns to the new data frame

df1 = pd.concat([df1, df.iloc[:, 16:26]], axis=1)

# save the new data frame to a new csv file

df1.to_csv('pdpu.csv', index=False)
```

```python
df_daiict = pd.read_csv('pdpu.csv')

df_daiictGrouped = df_daiict.groupby('profession').mean()

df_GroupedTransposed = df_daiictGrouped.T

# drop age column

df_GroupedTransposed = df_GroupedTransposed.drop('Age')

plt.figure(figsize=(20, 10))

df_GroupedTransposed.plot(kind='bar', figsize=(20, 10))

# rotate the x axis labels

plt.xticks(rotation=80)
plt.xlabel('Profession')
plt.ylabel('Average Rating')
plt.title('Average Rating of Heuristics in PDPU')
plt.show()

plt.savefig('daiict.png')
```

```python
df_daiict = pd.read_csv('pdpu.csv')    

bins = [0, 25, 30, 35, 100]
labels = ['<25', '25-30', '30-35', '35+']

df_daiict_age = pd.cut(df_daiict['Age'], bins=bins, labels=labels)

grouped_df = df_daiict.groupby(df_daiict_age).mean()

grouped_df.to_csv('pdpu_age.csv')

print(grouped_df)

# grouped_df = grouped_df.drop('Age')

# plot the grouped data

# drop column with index 1 

grouped_df = grouped_df.drop(grouped_df.columns[0], axis=1)

plt.figure(figsize=(20, 10))

grouped_df.plot(kind='bar', figsize=(20, 10))

# rotate the x axis labels


plt.xticks(rotation=80)

plt.xlabel('Age Group')

plt.ylabel('Average Rating')

plt.title('Average Rating of Heuristics in PDPU')

plt.show()
```
